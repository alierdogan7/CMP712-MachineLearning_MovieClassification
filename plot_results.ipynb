```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

results = {}
labels = ["Log.Reg.", "MultinomialNB", "kNN", "SVM"]
bow_1gram_acc = [0.51, 0.55, 0.39, 0.46] 
bow_2gram_acc = [0.56, 0.51, 0.38, 0.49] 
tfidf_1gram_acc = [0.63, 0.57, 0.52, 0.58]
tfidf_2gram_acc = [0.58, 0.43, 0.54, 0.59]

def plot_line(scores, tag):
    plt.plot(labels, scores, label=tag)#, 'o')
    for i,j in enumerate(scores):
        plt.annotate(str(j),xy=(i,j),xytext=(-7,7), textcoords='offset points')
    
figure(figsize=(10, 8), dpi=80)
plot_line(bow_1gram_acc, "Bag of Words, 1-gram")
plot_line(bow_2gram_acc, "Bag of Words, 2-gram")
plot_line(tfidf_1gram_acc, "TF-IDF, 1-gram")
plot_line(tfidf_2gram_acc, "TF-IDF, 2-gram")

# plt.axis([0, 5, -1, 100])

plt.xlabel('Method')
plt.ylabel('Accuracy')
plt.legend(loc='upper right', numpoints=1)
plt.title("Accuracies / Methods")
plt.show()

```
